### Basic Statistics
Pandas allows you to quickly inspect data and get summary statistics. This is useful for looking at the broad trends in the data before expending time with more involved techniques.

Create a statistical summary of the patient data.

```python
# The describe method provides basic stats. The result is a Data Frame.
dsummary = patientdata.describe()
dsummary
```

_It can be helpful to transpose the result, with `patientdata.describe().transpose()`. When transposed, each row is the statistics of a specific category._

_If you have data that is not numeric, you can get statistics on that information by using `patientdata.describe(include='all')`._

```python
dsummary.T
```

It is possible to call individual methods that form part of the `describe()` summary:

```python
# Find the maximum BMI
patientdata.BMI.max()
```

_Pandas provides other aggregation methods that can be used similarly. These include `min`, `max`, `mean`, `median`, `std`, `var`, `skew`, `mad`, `kurtosis`, `autocorr`, `nunique`, `sem`._

_You can also check if a column `hasnans`, `is_monotonic`, or `is_unique`._  

Jupyter provides methods for retrieving quantile values:

```python
patientdata.BMI.quantile(0.4)
```

or ranges:

```python
patientdata.BMI.quantile([.25, .50, .75])
```

_You can retrieve as many quantile points as desired by passing an iterable with values beween 0 and 1._

_Other summary methods, such as `corr` (which can be used to calculate the correlation between two continuous variables), are useful when trying to understand how variables of interest in the dataset related to one another._

### Visualizing Data
While numerical statistics are a powerful way to quantitatively describe distributions, they also have some limitations. With very large amounts of data, they should be complemented by visualizations. The human mind can ingest and process enormous amounts of visual data and it is easier to visually recognize patterns that may not appear in a numerical description (such as bimodality).

Jupyter provides a strong set of features for creating plots and graphs. Under the hood it leverages Matplotlib. The visualizations in Pandaas include traditional charts such as scatterplots, histograms, bar charts, box plots, and area plots; in addition to more specialized visualizations such as hexbin plots, scatter matrixes, RadViz, and others.

Create a histogram for the age column.

```python
patientdata.age.hist()
```

Plot the the relationship between age and BMI.

```python
patientdata.plot(x='age', y='BMI')
```

_In this dataset, age and BMI are not correlated which results in something of an unhelpful mess._

The default plot type is a line plot. Re-plot the same relationship as a scatter plot. _This removes the connecting lines between the data points._

Make a scatter plot for age and bmi relation.

```python
patientdata.plot(x='age', y='BMI', kind='scatter')
```

### Filtering Data
Pandas provides ways to create boolean "masks" that describe whether datapoints meet a criteria.

```python
gt30 = patientdata.BMI >= 30
lt50 = patientdata.BMI < 50
```

Apply the mask.

```python
patientdata[gt30 & lt50]
```

_Be careful not to use `and` here.  That will translate to anding the entire sequences together, instead of using a bitwise operation. This is a case where normal Python syntax does not work._

Retrieve specific columns of interest.

```python
patientdata.loc[gt30 & lt50, ['BMI', 'diabetes', 'currentSmoker', 'cigsPerDay', 'TenYearCHD']]
```

_By convention, you should be using .loc to find columns in a DataFrame, and .iloc to find rows. The second argument is a list of the columns you are interested in.._

Sort the items in the dataset by BMI.

```python
patientdata.set_index('BMI').sort_index().head()
```

Fetch patients who have diabetes.

```python
patientdata.set_index('diabetes').sort_index().loc[1]
```
