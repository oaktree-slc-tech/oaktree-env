# GPU Environment Tests
This notebook includes tests for GPU acceleration to see if CUDA is able to run within PyTorch and TensorFlow.

```python
# Test PyTorch and TensorFlow to ensure that CUDA and GPU acceleration are available
import torch
import tensorflow as tf

print('PyTorch CUDA Available: %s\nTensorFlow GPU Available: %s' 
      % (torch.cuda.is_available(), len(tf.config.list_physical_devices('GPU')) > 0))
```
