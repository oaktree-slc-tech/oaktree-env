# Python: Computational Language of Data
Python is a general purpose interpreted programming language that allows you to use imperative, object oriented, or functional coding styles.

* A Python program is a sequence of definitions and commands
  - definitions evaluated
  - commands executed by Python interpreter in a shell
  
* Commands (statement) instruct the interpreter to do something
  - Python code can be typed directly in a shell or stored in a file that is read into the shell and evaluated
  
* Programs manipulate data objects
  - objects have a type that defines the kinds of things that the program can do
  - objects are
    + scalar (cannot be subdivided)
    + non-scalar with internal structure that can be accessed
  - Examples:
    + `int`: integers
    + `float`: flating point numbers
    + `bool`: Boolean values (True/False)
    + `NoneType`: special and has one value, `None`
    + the type of a program can be evaluated using `type`
    
* Python has an extensive standard library and believes in "batteries included"
  - Built-in types and methods cover many use-cases
  - Additional functionality can be quickly installed using the Python package manager `pip`

## Control Flow: `while` and `for`

### `while`
```python
while <condition>:
    <expression>
    <expression>
```

* <condition> evaluates to a Boolean
* if <condition> is True, execute inner code
* check <condition> again
* repeat until <condition> is False

````python
# Example 1: Dynamic Input
n = input("You are in the Lost Forest\n****************\n****************\n :)\n****************\n****************\nGo left or right? ")
while n == "right" or n == "Right":
    n = input("You are in the Lost Forest\n****************\n******       ***\n  (╯°□°）╯︵ ┻━┻\n****************\n****************\nGo left or right? ")
print("\nYou got out of the Lost Forest!\n\o/")
````

_When creating conditional checks such as `n == 'right' or n == 'Right'`, it can be better to normalize the input to prevent an execssive number of branches. The code below uses a string method `lower` to ensure that the user's input is always lowercase before being compared to the options._

```python
while n.lower() == 'right':
    ...
```

````python
# Example 2: Exit Loop Under Controlled Circumstances
n = 0
while n < 5:
    print(n)
    n = n+1
````

### `for`

```python
for <variable> in range(<start>, <end>):
    <expression>
    <expression>
```

* each time through the loop, <variable> takes a value from the sequence
* first time, <variable> starts at the smallest value
* next time, <variable> gets the next value in the sequence
    
`range(start, stop, step)`: Helper method that can be used to generate sequences of numbers

* default values are `start=0` and `step=1`
* loop until value is `stop-1`

````python
# Example 1: Iterate over range 
for n in range(5):
    print(n)
````

````python
# Example 2: Create a running sum
mysum = 0
for i in range(10):
    mysum += i
    print('Internal, step=%s. Sum=%s' % (i, mysum))
    
print(mysum)
````

`break` can be used to exit a loop

* skips remaining expressions in code block
* exits only inntermost loop

````python
# Example 3: Create a running sum and terminate iteration
mysum = 0
break_iter = 5
print('Stop iteration if the sum equals %s' % break_iter)

for i in range(5, 11, 2):
    mysum += i
    print('Internal: %s' % mysum)
    if mysum == break_iter:
        break
        mysum += 1

print(mysum)
````

* Iteration in Python is not limited to values in a `range`, any "container" can be iterated over
  - Arrays: members of a list or tuple
  - Key/value pairs in a hashmap (dictionary)
  - Individual characters in a string

* Python includes helper methods to modify how iteration occurs
  - `enumerate` can be used to obtain the index and value of a container: `enumerate(c)`

````python
# Example 4: Iterate over values in a container
s = ['alpha', 'beta', 'gamma', 'delta', 'epsilon']

for i, v in enumerate(s):
    print('Index: %s. Value: %s' % (i, v))
````

````python
# Example 5: Iterate over key and value in a dictionary
s = { 'a': 'alpha', 'b': 'beta', 'c': 'gamma', 'd': 'delta', 'e': 'epsilon' }
for k, v in s.items():
    print('English (key): %s. Greek (value): %s.' % (k, v))
````

## Exercise: House Hunting
You have just accepted a job in the San Francisco Bay area and you want to (eventually) buy a home. Since housing prices are very high in the Bay Area, you want to plan your savings and determine how long you will need to save before you can afford a down payment. To model a few different, you decide to create a simple program. The assumptions you are currently making are:

1. The value of the dream host should be set when the program starts (`total_cost`)
2. Portion of the cost needed for a down payment will also be constant (`portion_down_payment`) and can be set to 25% of the value of the home (`0.25`)
3. The amount you have currently saved can be modeled in a single variable (`current_savings`)
4. Assuming that you invest your current savings wisely, with annual return of r (in other words, at the end of each month, you receive an additional `current_savings*r/12` payment into your savings as interest). Assume that your investments earn a return of r=0.04 (4%).
5. Assume the annual salary is annual salary.
6. Assume you are going to dedicate a certain amount of your salary each month to saving fo rhte down payment. Call that `portion_saved`.
7. At the end of the month, the savings will increase by the return on your investment plus a percentage of your monthly salary (annual salary / 12).

Your program should ask for the following variables:

1. The starting annnual salary (`annual_salary`)
2. The portion of the salary to be saved (`portion_saved`)
3. The cost of the dream home (`total_cost`)

#### Hints
1. Organize your program into the following sections: input, process, iteration
2. Initialize some state variables to ensure that all values are available at the time you begin iterating

#### Test Case 1
* Annual Salary: $120,000
* Percent of salary: 0.1
* Number of months: 183

#### Test Case 2
* Annual Salary: $80,000
* Percent of salary: 0.1
* Number of months: 105

````python
# Exercise: House Hunting

# Set initial state variables
portion_down_payment = 0.25   # Percent of home value for down payment
r = 0.04   # Rate of investment return

# Input salary, portion saved, and total cost of home
total_cost = float(input('How mucn is the home: '))
annual_salary = float(input('Please enter the annual salary of the job: '))
portion_saved = float(input('Enter the amount to be saved each month as ratio (example: 0.1 for 10%)'))

# Monthly salary and down payment
monthly_salary = annual_salary/12
down_payment = portion_down_payment*total_cost

# Months and current savings
months = 0
current_savings = 0

while current_savings <= down_payment:
    
    # Increase current savings using savings and return on investment
    current_savings += portion_saved*monthly_salary + current_savings*r/12
    months += 1

print('%s months required to save %s down payment for a %s dollary home with a salary of %s setting aside %s'
     % (months, down_payment, total_cost, annual_salary, portion_saved))
````

#### Implementation Notes
* When I structure my programs, I like to put variables which may be adjusted in separate exections (but are nonetheless hard-coded) at the top. This makes them easier to find and modify.
* I chose to use a while loop and increment the months until the current savings are greater (or equal to) the down payment. There are other approaches that would likewise work.
* I used a "template string" (which is a string with placeholders) and merge with a data context (tuple containing values) to report the number of months and other variables of interest.
